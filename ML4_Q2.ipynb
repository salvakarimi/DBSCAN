```python
import random
from random import randint
import math
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas_profiling as pp
from sklearn import metrics #Import scikit-learn metrics module for accuracy calculation
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler
import matplotlib.cm as cm
import matplotlib.colors as colors
```

```python
compound = pd.read_csv('/content/Compound.txt',delim_whitespace=True,header=None,names=['x','y','c'])
d31 = pd.read_csv('/content/D31.txt',delim_whitespace=True,header=None,names=['x','y','c'])
spiral = pd.read_csv('/content/spiral.txt',delim_whitespace=True,header=None,names=['x','y','c'])
```

```python
compound.head()
```

# plotting the given data

```python
colors=10*["r","g","c","m","y","pink",'purple','yellow','aqua',"grey","indianred","cadetblue","salmon","sienna","dodgerblue","yellowgreen","skyblue","khaki","plum","gold","orchid","lime","b",'k']
```

```python
x1=compound['x']
y1=compound['y']
c1=compound['c']
for i in range(len(c1)):
  plt.scatter(x1[i],y1[i],color=colors[c1[i]])
plt.xlabel("X")
plt.ylabel("Y")
```

```python
x2=d31['x']
y2=d31['y']
c2=d31['c']
for i in range(len(c2)):
  plt.scatter(x2[i],y2[i],color=colors[c2[i]],s=18)
plt.xlabel("X")
plt.ylabel("Y")
```

```python
x3=spiral['x']
y3=spiral['y']
c3=spiral['c']
for i in range(len(c3)):
  plt.scatter(x3[i],y3[i],color=colors[c3[i]],s=18)
plt.xlabel("X")
plt.ylabel("Y")
```

## using DBSCAN (noise is displayed with black color)

# Compound

```python
db_default = DBSCAN(eps = 1.45, min_samples = 4.45).fit(compound[['x','y']])
labels = db_default.labels_
#print(labels)

#confusion matrix
confusion_matrix = metrics.cluster.contingency_matrix(c1, labels)
#purity score
print("purity score: "+str(np.sum(np.amax(confusion_matrix, axis=0)) / np.sum(confusion_matrix)))

#plotting ther esult
cvec = [colors[label] for label in labels]
plt.scatter(compound['x'], compound['y'], c = cvec)
```

# D31

```python
db_default = DBSCAN(eps = 0.45, min_samples = 5.88).fit(d31[['x','y']])
labels = db_default.labels_
#print(labels)

#confusion matrix
confusion_matrix = metrics.cluster.contingency_matrix(c2, labels)
#purity score
print("purity score: "+str(np.sum(np.amax(confusion_matrix, axis=0)) / np.sum(confusion_matrix)))

#plotting ther esult
cvec = [colors[label] for label in labels]
plt.scatter(d31['x'], d31['y'], c = cvec)
```

# Spiral

```python
db_default = DBSCAN(eps = 3.45, min_samples = 4.45).fit(spiral[['x','y']])
labels = db_default.labels_
#print(labels)

#confusion matrix
confusion_matrix = metrics.cluster.contingency_matrix(c3, labels)
#purity score
print("purity score: "+str(np.sum(np.amax(confusion_matrix, axis=0)) / np.sum(confusion_matrix)))

#plotting ther esult
cvec = [colors[label] for label in labels]
plt.scatter(spiral['x'], spiral['y'], c = cvec)
```
